```python
#create seperate dataframes for each type of products:
bread = df_panel_2.loc[df_panel_2['ItemName'].str.contains('לחם')]
# bread
milk = df_panel_2.loc[df_panel_2['ItemName'].str.contains('חלב')]
# milk
eggs = df_panel_2.loc[df_panel_2['ItemName'].str.contains('ביצים')]
# eggs
cucambers = df_panel_2.loc[df_panel_2['ItemName'].str.contains('מלפפון')]
# cucambers
tomatoes = df_panel_2.loc[df_panel_2['ItemName'].str.contains('עגבני')]
# tomatoes
potatoes = df_panel_2.loc[df_panel_2['ItemName'].str.contains('תפוח אדמה')]
# potatoes
lettuce = df_panel_2.loc[df_panel_2['ItemName'].str.contains('חסה')]
# lettuce
apples = df_panel_2.loc[df_panel_2['ItemName'].str.contains('תפוח עץ')]
# apples
bananas = df_panel_2.loc[df_panel_2['ItemName'].str.contains('בננות')]
# bananas
rice = df_panel_2.loc[df_panel_2['ItemName'].str.contains('אורז')]
# rice
pasta = df_panel_2.loc[df_panel_2['ItemName'].str.contains('פסטה')]
# pasta
tuna = df_panel_2.loc[df_panel_2['ItemName'].str.contains('טונה')]
# tuna
cottage_cheese = df_panel_2.loc[df_panel_2['ItemName'].str.contains('קוטג')]
# cottage_Cheese
hard_cheese = df_panel_2.loc[df_panel_2['ItemName'].str.contains('גבינה צהובה')]
# hard Cheese
```

<div style ='background:lightgreen;dont-size:15px'> in order to see the change of prices,lets see which item code appears the most in each group 


```python
#convert the dataframe to series 
headerNames = ["ItemCode", "store_id", "upload_date","PriceUpdateDate","ItemName","ManufacturerName","UnitQty",
"ChainName","SubChainId","SubChainName","StoreId","StoreName","Address","City","ZipCode","ItemPrice"]
bread.columns = headerNames
milk.columns = headerNames
eggs.columns = headerNames
cucambers.columns = headerNames
tomatoes.columns = headerNames
potatoes.columns = headerNames
lettuce.columns = headerNames
apples.columns = headerNames
bananas.columns = headerNames
rice.columns = headerNames
pasta.columns = headerNames
tuna.columns = headerNames
cottage_cheese.columns = headerNames
hard_cheese.columns = headerNames
```
